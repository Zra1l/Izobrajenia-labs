```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from skimage.metrics import structural_similarity
```

```python
gray_image = cv2.imread('image\sar_1_gray.jpg',0)
```

```python
plt.imshow(gray_image,cmap='gray') # без  cmap='gray' цвета могут быть некоректными
```

```python
histSize = 256# количество бинов в гистограмме (0-255)
histRange = (0, 256)# диапазон значений яркости
accumulate = False# не накапливать гистограмму между вызовами

b_hist = cv2.calcHist([gray_image], [0], None, [histSize], histRange, accumulate=accumulate)# ну вместо b gray_image подставил и всё)
```

```python
plt.plot(b_hist)
```

```python
gamma = 0.2
output_image = 255 * (gray_image / 255) ** gamma
plt.imshow(output_image,cmap='gray')
```

```python
mse = mean_squared_error(gray_image, output_image)
ssim = structural_similarity(gray_image, output_image, data_range=255)

print(f"MSE: {mse}")
print(f"SSIM: {ssim}")
```

```python
mean = gray_image.mean()
std = gray_image.std()

eq_gray_image = cv2.equalizeHist(gray_image)

mean_eq_gray_image = eq_gray_image.mean()
std_eq_gray_image = eq_gray_image.std()

corrected_image = mean_eq_gray_image + (gray_image.astype(np.int16) - mean) * (std_eq_gray_image / std)
corrected_image = np.clip(corrected_image, 0, 255).astype(np.uint8)

plt.imshow(corrected_image, cmap='gray')
```

```python
_, thresh1 = cv2.threshold(gray_image, 124, 255, cv2.THRESH_BINARY)# _ отбрасывает первый возращаемый параметр
_, thresh2 = cv2.threshold(gray_image, 50, 255, cv2.THRESH_BINARY) 
_, thresh3 = cv2.threshold(gray_image, 150, 255, cv2.THRESH_BINARY)
# if пиксель >= порог:  пиксель = 255 (белый) else: пиксель = 0 (черный)
```
