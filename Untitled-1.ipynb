```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from skimage.metrics import structural_similarity
```

```python
gray_image = cv2.imread('sar_1_gray.jpg',0)
```

```python
plt.imshow(gray_image,cmap='gray')
```

```python
histSize = 256# количество бинов в гистограмме (0-255)
histRange = (0, 256)# диапазон значений яркости
accumulate = False# не накапливать гистограмму между вызовами

b_hist = cv2.calcHist([gray_image], [0], None, [histSize], histRange, accumulate=accumulate)# ну вместо b gray_image подставил и всё)
```

```python
plt.plot(b_hist)
```

```python
gamma = 0.2
output_image = 255 * (gray_image / 255) ** gamma
plt.imshow(output_image,cmap='gray')
```

```python
mse = mean_squared_error(gray_image, output_image)
ssim = structural_similarity(gray_image, output_image, data_range=255)

print(f"MSE: {mse}")
print(f"SSIM: {ssim}")
```

```python
cdf = b_hist.cumsum()#тут уже коментарий надо ([2, 5, 3, 1] - представим что оригинал, а вот итог  [2, 7, 10, 11] (2, 2+5=7, 7+3=10, 10+1=11))
cdf_normalized = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min()) # ну в инете написано что без умножения на 255 значения будут 0 - 1 - а нам такое не подойдёт
cdf_normalized = cdf_normalized.astype('uint8')# - для целых чисел
stat_corrected = cdf_normalized[gray_image]# ну тут заменяем значения для грей имаге на значения cdf_normalized скорекитированого
plt.imshow(stat_corrected, cmap='gray') # а тут показываем
```
